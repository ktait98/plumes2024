```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 30)
```

```python
fl_df = pd.read_csv("fl.csv")
```

```python
plume_df = pd.read_csv("plume_evolution.csv")
plume_df.head(200)
```

```python
# merge dataframes on waypoint and time
concs_df = pd.merge(fl_df, plume_df, on=['waypoint'])
concs_df = concs_df[['waypoint', 'time_x', 'time_y', 'flight_id', 'longitude_y', 'latitude_y', 'co2_m', 'h2o_m', 'so2_m', 'nox_m', 'co_m', 'hc_m', 'nvpm_m','sigma_yy', 'width', 'depth', 'age', 'lon_plus', 'lat_plus', 'lon_minus', 'lat_minus', 'advection_distance']].

concs_df[['time_y', 'waypoint', 'longitude_y', 'latitude_y', 'lon_plus', 'lat_plus', 'lon_minus', 'lat_minus', 'width', 'advection_distance']]
```

```python
plt.figure()
plt.plot(concs_df["width"], label = 'width')
plt.plot(concs_df['advection_distance'], label = 'advection_distance')
```
